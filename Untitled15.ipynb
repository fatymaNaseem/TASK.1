```python
def process_csv(data):
    """
    Processes a given DataFrame by performing the following steps:
    - Checks for null values
    - Displays column info and descriptive statistics
    - Converts 'Date' column to datetime format

    Parameters:
    data (pd.DataFrame): The DataFrame to process

    Returns:
    pd.DataFrame: The processed DataFrame
    """
    # Check for null values, column info, and descriptive statistics
    print("\nChecking for null values:\n", data.isnull().sum())
    print("\nColumn Info:\n")
    print(data.info())
    print("\nDescriptive Statistics:\n", data.describe())

    # Convert 'Date' column to datetime
    if 'Date' in data.columns:
        data['Date'] = pd.to_datetime(data['Date'])
    else:
        print("No 'Date' column found in the data!")

    return data

# Process the uploaded data
processed_data = process_csv(data)
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to analyze Instagram reach data
def analyze_reach(data):
    """
    Analyzes Instagram reach data with visualizations and statistics.

    Parameters:
    data (pd.DataFrame): DataFrame containing 'Date' and 'Instagram reach' columns.

    Returns:
    pd.DataFrame: DataFrame with additional 'Day' column.
    """
    # Ensure the 'Date' column is in datetime format
    if 'Date' not in data.columns or 'Instagram reach' not in data.columns:
        raise ValueError("The dataset must contain 'Date' and 'Instagram reach' columns.")

    if not pd.api.types.is_datetime64_any_dtype(data['Date']):
        data['Date'] = pd.to_datetime(data['Date'])

    # Line Chart
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Instagram reach'], marker='o', linestyle='-', color='b')
    plt.title('Instagram Reach Over Time')
    plt.xlabel('Date')
    plt.ylabel('Instagram Reach')
    plt.grid(True)
    plt.show()

    # Box Plot
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=data['Instagram reach'], palette='coolwarm')
    plt.title('Distribution of Instagram Reach')
    plt.ylabel('Instagram Reach')
    plt.show()

    # Add Day Column
    data['Day'] = data['Date'].dt.day_name()

    # Group by Day of the Week
    day_stats = data.groupby('Day')['Instagram reach'].agg(['mean', 'median', 'std']).reindex(
        ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    )
    print("\nDay-wise Statistics:\n", day_stats)

    # Bar Chart for Days of the Week
    plt.figure(figsize=(10, 6))
    sns.barplot(x=day_stats.index, y=day_stats['mean'], palette='plasma')
    plt.title('Average Instagram Reach by Day of the Week')
    plt.xlabel('Day of the Week')
    plt.ylabel('Average Reach')
    plt.xticks(rotation=45)
    plt.show()

    return data

# Example usage
# Replace 'your_file.csv' with your actual CSV file path
from google.colab import files
uploaded = files.upload()

# Load the data
for file_name in uploaded.keys():
    data = pd.read_csv(file_name)

# Call the function
processed_data = analyze_reach(data)

```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.api import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

def forecast_reach(data):
    """
    Forecast Instagram reach using a SARIMA model.

    Parameters:
    data (pd.DataFrame): DataFrame containing 'Date' and 'Instagram reach' columns.

    Returns:
    None
    """
    # Ensure the 'Date' column is in datetime format
    if 'Date' not in data.columns or 'Instagram reach' not in data.columns:
        raise ValueError("The dataset must contain 'Date' and 'Instagram reach' columns.")

    if not pd.api.types.is_datetime64_any_dtype(data['Date']):
        data['Date'] = pd.to_datetime(data['Date'])

    # Sort data by date (important for time series analysis)
    data = data.sort_values(by='Date')

    # Line Chart to Check Trends and Seasonal Patterns
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Instagram reach'], marker='o', linestyle='-', color='g')
    plt.title('Instagram Reach - Trends and Seasonal Patterns')
    plt.xlabel('Date')
    plt.ylabel('Instagram Reach')
    plt.grid(True)
    plt.show()

    # Autocorrelation and Partial Autocorrelation Plots
    plt.figure(figsize=(12, 6))
    plot_acf(data['Instagram reach'], lags=20)
    plt.show()

    plt.figure(figsize=(12, 6))
    plot_pacf(data['Instagram reach'], lags=20)
    plt.show()

    # SARIMA Model
    try:
        model = SARIMAX(data['Instagram reach'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
        results = model.fit(disp=False)
        print(results.summary())
    except Exception as e:
        print(f"Model fitting failed: {e}")
        return

    # Forecasting
    forecast_steps = 10
    forecast = results.get_forecast(steps=forecast_steps)
    forecast_index = pd.date_range(start=data['Date'].iloc[-1] + pd.Timedelta(days=1), periods=forecast_steps)
    forecast_values = forecast.predicted_mean

    # Plot Forecast
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Instagram reach'], label='Actual')
    plt.plot(forecast_index, forecast_values, label='Forecast', color='red', linestyle='--')
    plt.title('Instagram Reach Forecast')
    plt.xlabel('Date')
    plt.ylabel('Instagram Reach')
    plt.legend()
    plt.grid(True)
    plt.show()
```
